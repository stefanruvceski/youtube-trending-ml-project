```python
model = VGG16(weights='imagenet', include_top=False)
pkl_file  = 'C:\\Users\\Stefan\\Downloads\\US_trending.pkl'

df = pd.read_pickle(pkl_file)
images = []
i = 1
data = [];
for video_id,image_url in zip(df['video_id'],df['thumbnail_link']):
  response = requests.get(image_url)
  #img = Image.open(BytesIO(response.content))
  #img.save("images/{}".format(image_url))
  with Image.open(BytesIO(response.content)) as img:
        #img.save( "yt_thumbnails/image"+str(i)+".jpg")
        x = image.img_to_array(img)
        x = np.expand_dims(x, axis=0)
        x = preprocess_input(x)

        features = model.predict(x)
        data.append([video_id,image_url,features])
       
        i+=1
        
data 
```

```python
trdata = ImageDataGenerator()
traindata = trdata.flow_from_directory(directory="yt_thumbnails/train",target_size=(120, 90),class_mode=None,batch_size=30)
testdata = trdata.flow_from_directory(directory="yt_thumbnails/test",target_size=(120, 90),class_mode=None,batch_size=30)
```

```python
len(traindata)
```

```python
model = VGG16(weights='imagenet', include_top=False,input_shape=(120,90,3))
model.add(Dense(64, activation='relu'))
model.add(Dense(10, activation='softmax'))

model.compile(optimizer='adam',loss='categorical_crossentropy', metrics=['accuracy'])
model.fit_generator(traindata,steps_per_epoch=2,validation_data = testdata,validation_steps=2,epochs=5)
```
